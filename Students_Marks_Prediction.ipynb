```python
###storing the data of the marks of mid terms in different variablees specific tothe uses
dataCE1=data['6CE1M']

```

```python
## now the marks of 6CE1 are in the binary form and are stored in a list with name dataCE1
len(dataCE1)
```

```python
## now similarly converting the numbers obtained by the students in the RTU are converted in the binary form 
## Pass = 1 and Fail = 0
## these marks are stored in the list with name of data
dataCE1RTU=[]
for i in data['6CE1']:
    if i>26:
        dataCE1RTU.append(1)
    else:
        dataCE1RTU.append(0)
```

```python
len(dataCE1RTU)
```

```python
## converting the two lists dataCE1 and dataCE1RTU into single 2-D Numpy  array  for finding correlation between the data
#data_under_testing_6CE1 = np.column_stack((dataCE1,dataCE1RTU)) 
#print (pearsonr(dataCE1,dataCE1RTU))
dataCE1Numpy=np.asarray(dataCE1)
dataCE1RTUNumpy=np.asarray(dataCE1RTU)
print(dataCE1Numpy.size)
print(dataCE1RTUNumpy.size)
```

```python
dataCE1Numpy = dataCE1Numpy.reshape((123,1))
X=dataCE1Numpy
y=dataCE1RTUNumpy
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
logistic = linear_model.LogisticRegression()
logistic.fit(X_train,y_train)
```

```python
print(" ## training data size = ",len(X_train))
print(" ## Testing data size = ", len(X_test))
print(" ## training data size (y)= ",len(y_train))
print(" ## Testing data size (y) = ", len(y_test))
```

```python
from sklearn.svm import SVC
tr_model = SVC(kernel='linear', C=1.0)
tr_model.fit(X_train, y_train)
```

```python
### predicting the testing data using SVM
tr_predict_test = tr_model.predict(X_test)
from sklearn import metrics
print("Accuracy by  SVM = {}".format(metrics.accuracy_score(y_test, tr_predict_test)))

## predicting the training data i.e. to check if the data is being overfitted

### predict values using the training data
tr_predict_train= tr_model.predict(X_train)
from sklearn import metrics
print("accuracy train =  {0}".format(metrics.accuracy_score(y_train, tr_predict_train)))

```

```python
from sklearn.neural_network import MLPClassifier
tr_model = MLPClassifier()
tr_model.fit(X_train, y_train)
```

```python
### predicting the testing data
tr_predict_test = tr_model.predict(X_test)
from sklearn import metrics
print("Accuracy by  MLPClassifier = {}".format(metrics.accuracy_score(y_test, tr_predict_test)))
```

```python
## now we are testing if the student scores 8 marks in the midterm
result=tr_model.predict(1)
if result[0]==1:
    print("pass")
else:
    print("fail")
## this is used to calculate the probabilty that the student will be passed. If the percentage_predicted[2] >50 
## then the student is said to be passed
percentage_predicted=tr_model.predict_proba(1)
print("confidence score = ",percentage_predicted)
```

```python
percentage_predicted_array_numpy=[]
for i in range(10):
    percentage_predicted=tr_model.predict_proba(i)
    percentage_predicted_array_numpy.insert(i,percentage_predicted[:,1])
print(percentage_predicted_array_numpy)

## as we can see the percentage_predicted_array_numpy is an python list of numpy arrays hence we have to flatten it
## to make it usable as a list we have used np.concatenate(list of numpy array).ravel().tolist()
print()
print()
## as 
percentage_predicted_array = np.concatenate(percentage_predicted_array_numpy).ravel().tolist()
print(percentage_predicted_array)    
x_axis=[]
x_axis = [i for i in range(10)]
print(x_axis)
```

```python
plt.plot(percentage_predicted_array,x_axis)
plt.plot()
plt.ylabel("numbers obtained in mid terms out of 10")
plt.xlabel("probability score of passing in University exams ")

plt.show()
```

```python
## from the above line plot we observe that the students who score more than 2 marks are going to pass with 60% probality
## So we can infer that most of the students who score below  passing marks in mid terms get passed in university papers
## for that we have to analyze which factors are responcible for the poor perforamance of the students in the mid term exams
```
